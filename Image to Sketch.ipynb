<h3 ><center><b><i>Converting an image into pencil sketch</i></b></center></h3>

## Importing libraries

The only library required to convert an image into a pencil sketch with Python is an **OpenCV library** in Python.

```python
# to convert an image into a pencil sketch
import cv2

# to display images
import matplotlib.pyplot as plt
```

## Reading image

```python
image = cv2.imread("python.png")
```

## Show image

```python
# shows image in a new window
cv2.imshow("Original Image", image)
cv2.waitKey(0)
```

## Displaying image using matplotlib

```python
# displays image using matplotlib library in BGR format
plt.imshow(image)
plt.axis(False)
plt.show()
```

## Steps to convert an image into pencil sketch

###  Step 1 - Converting to RGB format

#### Using cvtColor method of OpenCV

```python
RGB_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_image)
plt.axis(False)
plt.show()
```

### Step 2 - Converting RGB image to Grey Image

#### Using cvtColor function of OpenCV.

```python
gray_image=cv2.cvtColor(RGB_image, cv2.COLOR_BGR2GRAY)
cv2.imshow("Grey Image", gray_image)
cv2.waitKey(0)
```

## Step 3 - Inverting the greyscale image

```python
inverted_image = 255 - gray_image
cv2.imshow("Inverted Image", inverted_image)
cv2.waitKey()
```

## Step 4 - Blur the image

#### Apply Gaussian blur to the image. 
- The second argument to the function is the kernel size, if should be a pair of odd numbers.
- Larger the kernel size, more blurred the image will be and it will lose its subtle features.

```python
blur_image = cv2.GaussianBlur(inverted_image, (21,21),0)
```

## Step 5 - Inverting Blurred Image

```python
invblur_image = 255 - blur_image
```

## Step 6 - Final Pencil Sketch

#### The sketch can be obtained by performing bit-wise division between the grayscale image and the inverted-blurred image.

```python
# creating pencil sketch
pencil_sketch = cv2.divide(gray_image, invblur_image, scale=256.0)

# displaying sketch using matplotlib
sketch=cv2.cvtColor(pencil_sketch, cv2.COLOR_BGR2RGB)
plt.imshow(sketch)
plt.axis('off')
plt.show()
```

## Original Image vs Pencil Sketch

```python
# displaying original image
plt.figure(figsize=(8,4))
plt.title('Original image', size=15)
plt.imshow(RGB_image)
plt.axis('off')
plt.show()

# displaying sketch image
plt.figure(figsize=(8,4))
plt.title('Pencil Sketch', size=15)
rgb_sketch=cv2.cvtColor(pencil_sketch, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```

Image has successfully been converted into Pencil sketch
